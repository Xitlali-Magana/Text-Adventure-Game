```python
# to control speed of game 
import time
import random

# what will happen when the player wins
def win():
    print("Congratulations! You won the game.")
    
# what will happen when the player loses
def fail():
    print("Oops, you failed the game. Better luck next time!")
    
    # try again, loop until succesfully answered  
    redo = input(prompt ="""
Do you want to try again?
1. Yes
2. No
>

""")
    if "1" in redo or "yes" in redo: # ensure either text or number goes through 
        time.sleep(1)
        stage1()
    # if player decides not to play again, this will appear
    else:
        print('Sorry to see you go')
    
    
# stage 1 description 
def stage1():
    # this is done to control the results of the keypad entry.
    # If they input the correct answer in the keypad, it will move to the next stage
    access_granted = False
    # description of stage 1
    print("""
Welcome to Stage 1!
You are at the Monsters Inc. factory.
You need to figure out how to access the door to the human world.""")
    # Question 1 
    code = input(prompt = """
    Do you talk to your colleague to obtain a code to open the door.
    1. Yes. They give you the code. However you will have to work to obtain screams. 
    2. No. You get a clue and have to guess the code. However, you automatically get screams.  
        >""")  
    # control for while loop, if player answers with something that isnt an option, question 1 will loop
    x = 1
    while x > 0:
        # Question 1 - answer 1
        if code.lower() == "yes" or code.lower() == "code" or code == "1": # ensure either text or number goes through 
            print("""
    The door to the human world, how frightful! The code to the door is \'123\'""")
            # Question 1.1
            keypad = input(prompt = """
    You walk to the door.
    A keypad can be seen to the left.
    Type in the code?
        >""")
            # Question 1.1 correct answer 
            if keypad == "123":
                print("""
    Yay! You have successfully accessed the door to the human world!
    """)
                # break loop and move onto stage 2
                x -= 1 
                stage2()
            # Question 1.1 incorrect answer
            # Loop back to question 1.1
            elif keypad!= "123":
                keypad = input(prompt = """
    That\'s incorrect. Try again.
        >""")
            # additional check, unlikely, but an error message might appear
            else:
                print("""
    Something went wrong.""")
        # Question 1 - answer 2
        elif code.lower() == "no" or code == "2": # ensure either text or number goes through 
            # list of possible codes
            code_lst = [123, 456, 789, 147, 258, 369, 159, 357, 761, 943]
            #description of the next input 
            print("""
    The code is 3 digits. 
    You will be given a list of potential codes that will open the door, choose wisely.""")
            # Question 1.2
            codelist = input(prompt = """
    Which list of potential codes would you like to be shown?
    1. A list of 10 potential codes, but only 5 attempts at opening the door.
    2. A list of 5 potential codes, but only 2 attempts at opening the door.
        >""")
            # Question 1.2 - answer 1
            if codelist == "1" or codelist == "10":
                # printing all 10 potential codes
                print(code_lst)
                # Question 1.2.1
                code_clue = input(prompt = """
    A keypad can be seen to the left, you have 5 attempts.
    What is the code?
        >""")
                # control for nested while loop 
                attempts = 0
                # only 5 attempts at question 1.2.1
                while attempts < 5:
                    # Question 1.2.2 correct answer
                    if code_clue == "123":
                        print("""
    Yay! You have successfully accessed the door to the human world!
    And collected screams! 
    You get to skip stage 2 and go straight to delivering the screams!
    """)
                        time.sleep(2)
                        # break while loop 
                        access_granted = True
                        break
                    # Question 1.2.2 incorrect answer
                    elif code_clue != "123":
                        # reduce number of attempts
                        attempts += 1
                        print(f"""
    That\'s incorrect. You have {5-attempts} attempts remaining.""")
                        # Redo question 1.2.2 until attempts = 0
                        code_clue = input(prompt = """
    What is the code?
            >""")
                    # additional security check, unlikely 
                    else: 
                        print("""
    Something went wrong.""")
            # Question 1.2 - answer 2
            elif codelist == "2" or codelist == "5":
                # opening only the first 5 potential codes
                for i in range(4):
                    print(random.choice(code_lst))
                print(code_lst[0])
                    ################# #print(code_lst[i])
                # Question 1.2.2
                code_clue = input(prompt = """
    A keypad can be seen to the left, you have 2 attempts.
    What is the code?
        >""")
                # control for nested while loop 
                attempts = 0
                access_granted = False
                # only 2 attempts at question 1.2.2
                while attempts < 2:
                    # question 1.2.2 correct answer
                    if code_clue == "123":
                        print("""
    Yay! You have successfully accessed the door to the human world!
    And collected screams! 
    You get to skip stage 2 and go straight to delivering the screams!
    """)
                        time.sleep(2)
                        access_granted = True
                        break
                    # question 1.2.2 incorrect answer
                    elif code_clue != "123":
                        # reduce number of attempts
                        attempts += 1
                        print(f"""
    That\'s incorrect. You have {2-attempts} attempts remaining.""")
                        # redo question 1.2.2 until attempts = 0
                        code_clue = input(prompt = """
    What is the code?
            >""")
                    # check, unlikely to occur
                    else:
                        print("""
    Something went wrong.""") 
            else: 
                print("""
    Something went wrong.""")
            # check to break while loop             
            if access_granted == True:
                time.sleep(2)
                # break loop, move onto next stage 
                x -= 1 
                stage3() 
            else:
                # loop breaks, but game is failed
                fail()
                break
        # redo question 1, loop until successfully answered
        else:
            print("""
    Invalid action. Try again.""")
            # Redo question 1.2, loop until successfully answered
            code = input(prompt = """
    Do you talk to your colleague to obtain a code to open the door.
    1. Yes. They give you the code. However you will have to work to obtain screams. 
    2. No. You get a clue and have to guess the code. However, you automatically get screams.  
        >""")

# description of stage 2
def stage2():
    print("""
    

Welcome to Stage 2!
You are in a human child's bedroom.
Your objective is to scare the child and collect enough screams to power the factory.
Avoid obstacles and the child's parents.""")
    # while loop control 
    screams = 0
    while screams < 10:
        #Question 2
        action = input(prompt = """
    What action do you take?
    1. Hide and figure out a plan of action
    2. Scare the child
    3. Get out
        >""")
        # Question 2 - answer 1
        if "hide" in action.lower() or "1" in action: # ensure either text or number goes through 
            print("""
    You have successfully hidden from the parents.""")
            time.sleep(2)
            print("""
    The child's parents came in and found you.""")
            # fail game, break loop 
            fail()
            break 
        # Question 2 - answer 2
        elif "scare" in action.lower() or "2" in action: # ensure either text or number goes through 
            # add screams to break loop 
            screams += 1
            print("""
    Scream collected. Total screams:""", screams)
            print("""
    You have collected enough screams! Moving on to the next stage.""")
            # move onto stage 3
            stage3()
        # Question 2 - answer 3
        elif "get" in action.lower() or "3" in action: # ensure either text or number goes through 
            print("""
    The child's parents came in and found you as you ran towards the exit!""")
            # fail game, break loop 
            fail()
            break
        # Redo question 2, loop until successfully answered
        else:
            print("""
    Invalid action. Try again.""")

# desciption of stage 3
def stage3():
    print("""
Welcome to Stage 3!
You are back at the Monsters Inc. factory.
Your objective is to deliver the screams to the power plant as fast as possible.
Avoid obstacles and other monsters.
""")
    # Question 3
    delivery = input(prompt = """
    There are two routes to deliver the screams, which do you take?
    1. A faster route, but you will encounter obstacles.
    2. A slower route, but very safe.
        >""")
    # while loop control 
    y = 1
    while y > 0:
        # Question 3 - answer 1
        if "faster" in delivery.lower() or "1" in delivery: # ensure either text or number goes through 
            print("""
    You begin walking, luckily you haven't encountered any obstacles.
    
    
    
    Suddenly, you see 3 monsters approaching.""")
            # Question 3.1
            obstacle1 = input(prompt = """
    What do you do?
    1. Turn around
    2. Hide
    3. Walk towards them
        >""")
            # Question 3.1 - answer 1 
            if "turn" in obstacle1.lower() or "1" in obstacle1: # ensure either text or number goes through 
                print("""
    You turned around and were forced to take the long path.
    You were unable to deliver the screams in time""")
                # breaking while loop 
                y -= 1 
                # fail game
                fail()
                break
            # Question 3.1 - answer 2
            elif "hide" in obstacle1.lower() or "2" in obstacle1: # ensure either text or number goes through 
                print("""
    You hid and the monsters passed by without noticing you.
    You continue walking.
    
    You encounter another group of monsters.""")
                # Question 3.1.1
                obstacle2 = input(prompt = """
    Where do you hide?
    1. A vent. 
    2. A room.
        >""")
                # Question 3.1.1 - answer 1 
                if "vent" in obstacle2.lower() or "1" in obstacle2: # ensure either text or number goes through 
                    print("""
    You hide in the vent.
    To your suprise it is a shortcut to the screams delivery location.""")
                    # break the while loop 
                    y -= 1 
                    # win the game
                    win()
                # Question 3.1.1 - answer 2
                elif "room" in obstacle2.lower() or "2" in obstacle2: # ensure either text or number goes through 
                    print("""
    You hide in the room and wait for the monsters to passby. 
    You remain undetected.
    
    You come out and continue walking towards the screams delivery location.""")
                    # Question 3.1.1.1
                    obstacle3 = input(prompt = """
    You are running out of time to turn in the screams.
    What do you do?
    1. RUN
    2. Look for a shortcut to the screams delivery location.""")
                    # Question 3.1.1.1 - answer 1
                    if "run" in obstacle3.lower() or "1" in obstacle3: # ensure either text or number goes through 
                        print("""
    You run as fast as possible.
    You turn in the screams just in time!""")
                        # break the while loop 
                        y -= 1 
                        # win the game 
                        win()
                    # Question 3.1.1.1 - answer 1
                    elif "shortcut" in obstacle3.lower() or "2" in obstacle3: # ensure either text or number goes through 
                        print("""
    You search for a shortcut.
    Sadly, you run out of time to delivery the screams.""")
                        # break the while loop
                        y -= 1 
                        # fail the game
                        fail()
                        break
                    # Redo question 3.1.1.1, loop until successfully answered
                    else:
                        obstacle3 = input(prompt = """
    Invalid action. Try again.
    Hit any key to continue.""")
                # Redo question 3.1.1, loop until answered successfully 
                else:
                    obstacle2 = input(prompt = """
    Invalid action. Try again.
    Hit any key to continue.""")
            # Question 3.1 - answer 3
            elif "walk" in obstacle1.lower() or "3" in obstacle1: # ensure either text or number goes through 
                print("""
    You walk toward them and they take the screams away from you.""")
                # break while loop
                y -= 1 
                # fail the game
                fail()
                break
            # Redo question 3.1, loop until successfully answered
            else:
                obstacle1 = input(prompt = """
    Invalid action. Try again.
    Hit any key to continue.""")
        # Question 3 - answer 2
        elif "slower" in delivery.lower() or "2" in delivery: # ensure either text or number goes through 
            print("""
    You begin walking.
    You walk for a while with no obstacles.
    
    Suddenly, you turn the corner to the screams delivery location and see someone!
    Its your bestfriend Mike Wazowski!
    """)
            # Question 3.2
            friend = input(prompt = """
    What do you tell him?
        >""")
            print("""
    Hey friend! Its good seeing you, but i have to run!
    I need to finish my paperwork before Roz kills me!
    
    

    You continue walking and deliver the screams on time!""")
            # break the loop
            y -= 1 
            # win the game
            win()
        # Redo question 3, loop until succesfully answered 
        else:
            delivery = input(prompt = """
    Invalid action. Try again.
     
    There are two routes to deliver the screams, which do you take?
    1. A faster route, but the you will encounter obstacles.
    2. A slower route, but very safe.
        >""")
        
        
stage1()
```
