Welcome to the movie Monsters Inc! You are James P. Sullivan a monster also known as Sulley. Your job at Monsters Incorporated is to collect screams to power Monstropolis.
```python
import time
import random

# function to handle winning the game
def win():
    print("\nCongratulations! You won the game.")

# function to handle losing the game
def fail():
    print("\nOops, you failed the game. Better luck next time!")
    redo = input("Do you want to try again? (Yes/No) ")
    if redo.lower() in ['yes', 'y']:
        stage1()
    else:
        print('Sorry to see you go')

# stage 2
def stage1():
    print("\nWelcome to Stage 1!\nYou are at the Monsters Inc. factory.\nYou need to figure out how to access the door to the human world.")
    access_granted = False
##### Question 1 
    code = input("\nDo you talk to your colleague to obtain a code to open the door?\n1. Yes\n2. No\n> ")
    while True:
##### Question 1 - answer 1
        if code.lower() in ["yes", "code", "1"]:
            print("\nThe door to the human world, how frightful! The code to the door is '123'")
##### Question 1.1
            keypad = input("\nYou walk to the door. A keypad can be seen to the left. Type in the code? > ")
##### Question 1.1 correct answer
            if keypad == "123":
                print("\nYay! You have successfully accessed the door to the human world!")
                access_granted = True
                stage2()
                break
##### Question 1.1 incorrect answer
            else:
                print("\nThat's incorrect. Try again.")
##### Question 1 - answer 2
        elif code.lower() in ["no", "2"]:
            code_lst = [123, 456, 789, 147, 258, 369, 159, 357, 761, 943]
            print("\nThe code is 3 digits. You will be given a list of potential codes that will open the door, choose wisely.")
##### Question 1.2
            codelist = input("\nWhich list of potential codes would you like to be shown?\n1. A list of 10 potential codes, but only 5 attempts at opening the door.\n2. A list of 5 potential codes, but only 2 attempts at opening the door.\n>")
##### Question 1.2 - answer 1
            if codelist in ["1", "10"]:
                print(code_lst)
##### Question 1.2.1
                code_clue = input("\nA keypad can be seen to the left, you have 5 attempts. What is the code?\n> ")
                attempts = 0
                while attempts < 4:
                    if code_clue == "123":
                        print("\nYay! You have successfully accessed the door to the human world! And you skipped stage 2!")
                        time.sleep(2)
                        access_granted = True
                        stage3()
                        break
##### Question 1.2.2 incorrect answer
                    else:
                        attempts += 1
                        print(f"\nThat's incorrect. You have {5 - attempts} attempts remaining.")
                        code_clue = input("\nWhat is the code?\n> ")
                if access_granted == False:  
                    fail()
                    break
##### Question 1.2 - answer 2
            elif codelist in ["2", "5"]:
                for i in range(4):
                    print(random.choice(code_lst))
                print(code_lst[0])
##### Question 1.2.2
                code_clue = input("\nA keypad can be seen to the left, you have 2 attempts. What is the code?\n> ")
                attempts = 0
                access_granted = False
                while attempts < 2:
##### question 1.2.2 correct answer
                    if code_clue == "123":
                        print("\nYay! You have successfully accessed the door to the human world! And you skipped stage 2!")
                        time.sleep(2)
                        access_granted = True
                        stage3()
                        break
 ##### question 1.2.2 incorrect answer
                    else:
                        attempts += 1
                        print(f"That's incorrect. You have {2 - attempts} attempts remaining.")
                        code_clue = input("\nWhat is the code?\n> ")
                if access_granted == False:
                    fail()
                    break
        else: 
            code = input("\nDo you talk to your colleague to obtain a code to open the door?\n1. Yes\n2. No\n> ")

# stage 2
def stage2():
    time.sleep(2)
    print("\nWelcome to Stage 2!\nYou are in a human child's bedroom.\nYour objective is to scare the child and collect enough screams to power the factory.\nAvoid obstacles and the child's parents.")
    screams = 0
##### Question 2
    action = input("\nWhat action do you take?\n1. Hide and figure out a plan of action\n2. Scare the child\n3. Get out\n> ")
    while screams < 1:
##### Question 2 - answer 1   
        if action.lower() in ["hide", "1"]:
            print("\nYou have successfully hidden from the parents.")
            time.sleep(2)
            print("\nThe child's parents came in and found you.")
            fail()
            break 
        elif action.lower() in ["scare", "2"]:
            screams += 1
            time.sleep(0.5)
            print(f"Scream collected.\nTotal screams: {screams}")
            if screams <= 1:
                print("\nYou have collected enough screams! Moving on to the next stage.")
                stage3()
                break
        elif action.lower() in ["get", "3"]:
            print("\nThe child's parents came in and found you as you ran towards the exit!")
            fail()
            break
        else:
            action = input("\nWhat action do you take?\n1. Hide and figure out a plan of action\n2. Scare the child\n3. Get out\n> ")    
            
# stage 3
def stage3():
    time.sleep(2)
    print("\nWelcome to Stage 3!\nYou are back at the Monsters Inc. factory.\nYour objective is to deliver the screams to the power plant as fast as possible.\nYou are back at the Monsters Inc. factory.")
    y = 0
##### Question 3 
    delivery = input("\nThere are two routes to deliver the screams, which do you take?\n1. A faster route, but you will encounter obstacles.\n2. A slower route, but very safe.\n> ")
    while y < 1:
        if delivery.lower() in ["faster", "1"]:
            print("\nYou begin walking, luckily you haven't encountered any obstacles.")
            print("\nSuddenly, you see 3 monsters approaching.")
            obstacle1 = input("\nWhat do you do?\n1. Turn around\n2. Hide\n3. Walk towards them\n> ")
            z = 0
            while z < 1:
##### Question 3.1 answer 1
                if obstacle1.lower() in ["turn", "1"]:
                    print("\nYou turned around and were forced to take the long path.\nYou were unable to deliver the screams in time")
                    y += 1
                    z += 1
                    fail()
                    break
##### Question 3.1 answer 2
                elif obstacle1.lower() in ["hide", "2"]:
                    print("\nYou hid and the monsters passed by without noticing you.\nYou continue walking.\nYou encounter another group of monsters.")
##### Question 3.1.1
                    obstacle2 = input("\nWhere do you hide?\n1. A vent.\n2. A room.\n> ")
                    x = 0
                    while x < 1:
##### Question 3.1.1 answer 1
                        if obstacle2.lower() in ["vent", "1"]: 
                            print("\nYou hid in the vent.\nTo your suprise it is a shortcut to the screams delivery location.")
                            y += 1
                            z += 1
                            x += 1
                            win()
                            break
##### Question 3.1.1 answer 2
                        elif obstacle2.lower() in ["room", "2"]: 
                            print("\nYou hid in the room and wait for the monsters to passby.\nYou remain undetected.\nYou come out and continue walking towards the screams delivery location.")
##### Question 3.1.1.1
                            obstacle3 = input("\nYou are running out of time to turn in the screams.\nWhat do you do?\n1. RUN\n2. Look for a shortcut to the screams delivery location.\n> ")
                            a = 0
                            while a < 1:
##### Question 3.1.1.1 answer 1
                                if obstacle3.lower() in ["run", "1"]:
                                    print("\nYou ran as fast as possible.\nYou turn in the screams just in time!")
                                    y += 1
                                    z += 1
                                    x += 1
                                    a += 1
                                    win()
                                    break
##### Question 3.1.1.1 answer 2
                                elif obstacle3.lower() in ["shortcut", "2"]:
                                    print("\nYou search for a shortcut.\nSadly, you run out of time to delivery the screams.")
                                    y += 1
                                    z += 1
                                    x += 1
                                    a += 1
                                    fail()
                                    break
                                else:
                                    obstacle3 = input("\nYou are running out of time to turn in the screams.\nWhat do you do?\n1. RUN\n2. Look for a shortcut to the screams delivery location.\n> ")
                        else:
                            obstacle2 = input("\nWhere do you hide?\n1. A vent.\n2. A room.\n> ")
##### Question 3.1 - answer 3
                elif obstacle1.lower() in ["walk", "3"]:
                    print("\nYou walk toward them and they take the screams away from you.")
                    y += 1 
                    z += 1
                    fail()
                    break
                else:
                    obstacle1 = input("\nWhat do you do?\n1. Turn around\n2. Hide\n3. Walk towards them\n> ")
##### Question 3 - answer 2
        elif delivery.lower() in ["slower", "2"]: 
            print("\nYou begin walking.\nYou walk for a while with no obstacles.")
            time.sleep(1)
            print("\nSuddenly, you turn the corner to the screams delivery location and see someone!\nIts your bestfriend Mike Wazowski!")
##### Question 3.2
            friend = input(prompt = "\nWhat do you tell him?\n> ")
            print("\nHey friend! Its good seeing you, but i have to run!\nI need to finish my paperwork before Roz kills me!")
            time.sleep(1)
            print("\nYou continue walking and deliver the screams on time!")
            y += 1 
            win()
            break
        else:
            delivery = input("\nThere are two routes to deliver the screams, which do you take?\n1. A faster route, but you will encounter obstacles.\n2. A slower route, but very safe.\n> ")
 
stage1()
```
